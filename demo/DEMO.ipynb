```python
# Installation:
# pip install palette-diagram

import palette_diagram as pt
import pandas as pd

df = pd.read_csv('./dataset/example.csv')
```

## Circular palette diagram

```python
pt.palette_diagram(df, index='data_id')
```

## Linear palette diagram

```python
pt.palette_diagram(df, palette_type='linear', index='data_id')
```

### Specific color assignments

```python
categories = ['category_B', 'category_A', 'category_C', 'category_D']

# The colors corresponding to 'category_A' and 'category_B' are flipped.
pt.palette_diagram(df, palette_type='linear', index='data_id', category_labels=categories)
```

### Specific color map

```python
pt.palette_diagram(df, palette_type='linear', index='data_id', cmap_name='Set2')
```
